creating a popularity-based recommendation system that uses the following criteria for filtering: the highest ratings, the highest number of users that have rated. 

```python
from math import*
import pandas as pd 
import numpy as np 
```

```python
movie_names = pd.read_csv("movies.csv")
movie_names.head()
```

```python
movie_ratings=pd.read_csv("ratings.csv")
movie_ratings.head()
```

```python
movie_data = pd.merge(movie_ratings,movie_names,on = "movieId")
movie_data.head()
```

```python
movie_data.groupby("title")["rating"].mean().head()
```

```python
movie_data.groupby("title")["rating"].mean().sort_values(ascending=False).head()
```

```python
ratings_mean_count = pd.DataFrame(movie_data.groupby('title')['rating'].mean())
ratings_mean_count
```

```python
movie_data[['title','rating']]
```

```python
ratings_mean_count['ratings_count']=pd.DataFrame(movie_data.groupby('title')['rating'].count())
ratings_mean_count.head()
```

```python
ratings_mean_count['rating']=round(ratings_mean_count['rating'],1)
ratings_mean_count.head()
```

```python
ratings_mean_count=ratings_mean_count[(ratings_mean_count['rating']>3) & (ratings_mean_count['ratings_count']>100)]
ratings_mean_count.head()
```
